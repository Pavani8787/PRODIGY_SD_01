```python
%%writefile app.py
import streamlit as st

st.set_page_config(page_title="Temperature Converter", page_icon="ðŸŒ¡ï¸")

st.title("ðŸŒ¡ï¸ Temperature Conversion Program")

# Text box input instead of increment box
temp_input = st.text_input("Enter the temperature value")

unit = st.selectbox("Select the unit", ["Celsius", "Fahrenheit", "Kelvin"])

if st.button("Convert"):
    try:
        temperature = float(temp_input)

        if unit == "Celsius":
            st.success(f"Fahrenheit: {(temperature * 9/5) + 32}")
            st.success(f"Kelvin: {temperature + 273.15}")

        elif unit == "Fahrenheit":
            celsius = (temperature - 32) * 5/9
            st.success(f"Celsius: {celsius}")
            st.success(f"Kelvin: {celsius + 273.15}")

        else:
            st.success(f"Celsius: {temperature - 273.15}")
            st.success(f"Fahrenheit: {((temperature - 273.15) * 9/5) + 32}")

    except ValueError:
        st.error("âŒ Please enter a valid numeric temperature value")
```
